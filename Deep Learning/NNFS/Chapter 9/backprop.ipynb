```python
import numpy as np
```

```python
x = [1.0, -2.0, 3.0] # input values
w = [-3.0, -1.0, 2.0] # weights
b = 1.0  # bias
```

```python
xw0 = w[0] * x[0]
xw1 = w[1] * x[1]
xw2 = w[2] * x[2]


z = xw0 + xw1 + xw2 + b  # 6.0
# ReLU
a = max(z, 0)  # 6.0

print(a)
```

```python
relu_dz = (z > 0)
```

**Suppose we got a derivative from next layer = 1**

```python
dvalue = 1
```

```python
drelu_dz = 1 * relu_dz
print(drelu_dz)
```

```python
dsum_dxw0 = 1

drelu_dxwo = drelu_dz * dsum_dxw0
drelu_dxwo
```

In such way we can apply it to all weights

```python
dsum_dxw1 = 1
drelu_dxw1 = drelu_dz * dsum_dxw1
drelu_dxw1
```

```python
dsum_dxw2 = 1
drelu_dxw2 = drelu_dz * dsum_dxw2
drelu_dxw2
```
