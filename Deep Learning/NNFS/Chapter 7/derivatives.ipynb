Just some beautiful visualizations for derivatives

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return 2 * x**2
```

```python
x_values = np.arange(0.0, 5.0, 0.01)
y_values = f(x_values)
```

```python
plt.plot(x_values, y_values)
plt.title("Quadratic function")
plt.show();
```

```python
def plot_tangent_line(x):
    y = f(x)
    
    p_val = 0.001
    x1 = x
    x2 = x + p_val
    y1 = y
    y2 = f(x2)

    derivative = (y2 - y1) / (x2 - x1)
    b = y2 - derivative * x2

    to_plot = [x1 - 0.9, x1, x1 + 0.9]

    plt.title("Derivative(-s)")
    plt.plot(x_values, y_values)
    plt.plot(x1, y1, "bo")
    plt.plot(to_plot, [derivative * x_t + b for x_t in to_plot], c="red")
    plt.show();
```

```python
plot_tangent_line(2)
```

```python
x_s = np.array([0, 1, 2, 3, 4, 5])
plot_tangent_line(x_s)
```
