```python
import pandas as pd
import numpy as np
```

```python
load_url = 'https://raw.githubusercontent.com/Yorko/mlcourse.ai/main/data/'
df = pd.read_csv(load_url + 'telecom_churn.csv')
df.head()
```

```python
print('shape: ', df.shape)
print('columns: ', df.columns)
```

```python
# general information about dataset:
df.info()
```

```python
# we can convert our column datatypes. Let's convert Churn from bool to int

df['Churn'] = df['Churn'].astype('int64')
df.info()
```

```python
# df.describe shows basical statistical characteristics for numerical features:

df.describe()
```

```python
# however we can also see stats for non-numerical data, by including it:

df.describe(include=['object', 'bool'])
```

```python
# there's also cool method value_counts for categorical data:

df['Churn'].value_counts()
```

```python
# we can also see it as a fraction:

df['Churn'].value_counts(normalize=True)
```

# Sorting

```python
# we can sort our dataframe by one of the variables:

# from lowest to highest with ascending=True (true by default)
df.sort_values(by='Total day charge').head()
```

```python
# from highest to lowest with ascending=False
df.sort_values(by='Total day charge', ascending=False).head()
```

```python
# but the craziest thing is that we can sort it by multiple columns:

df.sort_values(by=['Churn', 'Total day charge'], ascending=[True, False]).head()
```

# Indexing and retrieving data

```python
# we can index our dataset columns(features) with simple array-like indexing
df['Churn'].mean()
```

```python
# There's also boolean indexing when we select our values by condition.
print(df[df['Churn'] == 1]['Total day calls'].mean())
print(df[df['Churn'] == 0]['Total day calls'].mean())
```

```python
# and we can have many conditions in a row
df[(df['Churn'] == 0) & (df['International plan'] == 'No')]['Total intl minutes'].max()
```

```python
# there's also loc and iloc methods

# loc is used for indexing by name
# meaning our row is indexed with number and our column with it's name

# first five rows of churn column
df.loc[0:5, 'Churn']
```

```python
# iloc is used for indexing by numbers only

# first 5 rows of first 3 columns
df.iloc[0:5, 0:3]
```

```python
df[:1]
```

```python
df[-1:]
```

# Apply Functions to Cells, Columns, Rows

```python
# we use .apply() method to apply function to our data

df.apply(np.max)
```

```python
# states starting with letter t
df[df['State'].apply(lambda state: state[0] == 'T')].head()
```

```python
# .map() method is used for data replacement

d = {'No': 0, 'Yes': 1}

df['International plan'] = df['International plan'].map(d)
```

```python
df.head()
```

```python
# map is used for data changing
# apply is used for filtering
```

```python
# we can also use replace method (it's extremely simmilar to map, but...)

df = df.replace({"Voice mail plan": d})
df.head()
```

```python
# Replace ignores values not mentioned in dictionary d
# While map sets them to NaN
```

# Grouping

```python
df.head()
```

```python
columns_to_show = ["Total day minutes", "Total eve minutes", "Total night minutes"]

# grouping works as: df.groupby(column)[columns_to_show].function()
df.groupby(["Churn"])[columns_to_show].describe(percentiles=[])
```

# Summary Tables

```python
# we can build special crosstables to see distribution of categorical data

pd.crosstab(df['Churn'], df['International plan'])
```

```python
# and of course we can see only fractions

pd.crosstab(df['Churn'], df['International plan'], normalize=True)
```

```python
# we also have pivot tables

df.pivot_table(
    values=["Total day calls", "Total eve calls", "Total night calls"],
    index=['Area code'],
    aggfunc=np.mean
)

# where values are features we want to calculate statistics for
# index is a column we group by
# aggfunc is a function we apply (mostly statistical)
```

# **Let's try predicting Churn**

```python
# margin adds 'All' column and row
pd.crosstab(df['Churn'], df['International plan'], margins=True)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.countplot(x='International plan', hue='Churn', data=df);
```

```python
# we can see that churn rate is higher with International Plan
pd.crosstab(df['Churn'], df['International plan'], margins=True, normalize=True)
```

```python
# another important feature is customer calls

pd.crosstab(df['Churn'], df['Customer service calls'], margins=True)
```

```python
plt.figure(figsize=(8, 8))
sns.countplot(x='Customer service calls', hue='Churn', data=df);
```

```python
# we can see that our churn rate increases from 4 service calls

df['Many service calls'] = (df["Customer service calls"] > 3).astype("int")
df.head()
```

```python
pd.crosstab(df["Many service calls"], df["Churn"], margins=True)
```

```python
sns.countplot(x='Many service calls', hue='Churn', data=df);
```

```python
pd.crosstab(df["Many service calls"] & df['International plan'], df["Churn"], margins=True)
```
